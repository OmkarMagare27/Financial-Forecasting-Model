```python
import pandas as pd
import numpy as np
import os
import time
from dotenv import load_dotenv
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_pacf, plot_acf
from sklearn.metrics import mean_squared_error, mean_absolute_error
import tensorflow as tf
from pmdarima import auto_arima
from statsmodels.tsa.arima.model import ARIMA

# Load environment variables
load_dotenv()

# Default settings if environment variables are missing
Filename_address = os.getenv("FILE_ADDRESS", "/content/Dataset_01.csv")  # Default to "/content/Dataset_01.csv"
Output_address = os.getenv("OUTPUT_ADDRESS", "/content/")  # Default to "/content/"
close = "Adj_Close"
lag = int(os.getenv("LAG", 10))  # Default to 10
epochs = int(os.getenv("EPOCHS", 100))  # Default to 100
learning_rate = float(os.getenv("LEARNING_RATE", 0.001))  # Default to 0.001
batch_size = int(os.getenv("BATCH_SIZE", 32))  # Default to 32
number_nodes = int(os.getenv("NUMBER_NODES", 64))  # Default to 64
days = int(os.getenv("Prediction_days", 30))  # Default to 30
n = int(os.getenv("NN_LAGS", 5))  # Default to 5

# Load the data and make a dataframe
def data_loader():
    cols = ["Open", "High", "Low", "Close", "Adj_Close", "Volume"]
    data = pd.read_csv(Filename_address, index_col="Date", parse_dates=True)
    data.columns = cols
    data = data.dropna()
    print(f"The Shape of the Data-Set is : {data.shape}\nThe Data-Set is : \n{data.head()}\n")
    return data

# Plotting functions for various outputs
def plot_predictions(train, predictions, title):
    plt.figure(figsize=(10, 5))
    plt.plot(train.index, train, label='Actual')
    plt.plot(train.index, predictions, label='Predicted', color='red')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Close-Price')
    address = os.path.join(Output_address, title + ".jpg")
    plt.savefig(address)

def plot_raw_data(data):
    plt.figure(figsize=(10, 5))
    plt.plot(data.index, data[close], label='Close Price')
    plt.title('Raw Time Series Data')
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.legend()
    address = os.path.join(Output_address, 'Raw_Time_Series_Data.jpg')
    plt.savefig(address)

def plot_train_test(train, test):
    plt.figure(figsize=(10, 5))
    plt.plot(train.index, train, label='Train Set')
    plt.plot(test.index, test, label='Test Set', color='orange')
    plt.title('Train and Test Data')
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    address = os.path.join(Output_address, 'Train_and_Test_Data.jpg')
    plt.savefig(address)

def plot_prediction_errors(errors):
    plt.figure(figsize=(10, 5))
    plt.plot(errors, label='Prediction Errors')
    plt.title('Prediction Errors over Time')
    plt.xlabel('Time Step')
    plt.ylabel('Error')
    plt.legend()
    address = os.path.join(Output_address, 'Prediction_Errors_over_Time.jpg')
    plt.savefig(address)

def plot_final_predictions(test, final_predictions):
    plt.figure(figsize=(10, 5))
    plt.plot(test.index, test, label='Actual')
    plt.plot(test.index, final_predictions, label='Corrected Prediction', color='green')
    plt.title('Final Predictions with Error Correction')
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.legend()
    address = os.path.join(Output_address, 'Final_Predictions_with_Error_Correction.jpg')
    plt.savefig(address)

def plot_accuracy(mse, rmse, mae):
    metrics = ['MSE', 'RMSE', 'MAE']
    values = [mse, rmse, mae]
    plt.figure(figsize=(10, 5))
    plt.bar(metrics, values, color=['blue', 'orange', 'green'])
    plt.title('Model Accuracy Metrics')
    address = os.path.join(Output_address, 'Model_Accuracy_Metrics.jpg')
    plt.savefig(address)

def plot_arima_accuracy(mse, rmse, mae):
    metrics = ['MSE', 'RMSE', 'MAE']
    values = [mse, rmse, mae]
    plt.figure(figsize=(10, 5))
    plt.bar(metrics, values, color=['blue', 'orange', 'green'])
    plt.title('ARIMA Model Accuracy Metrics')
    address = os.path.join(Output_address, 'ARIMA_Model_Accuracy_Metrics.jpg')
    plt.savefig(address)

# Data partitioning for model development and training
def data_allocation(data):
    train_len_val = len(data) - days
    train, test = data[close].iloc[0:train_len_val], data[close].iloc[train_len_val:]
    print("\n--------------------------------- The Training Set is : -------------------------------------------\n")
    print(train)
    print(f"\nThe Number of Enteries : {len(train)}\n")
    print("\n--------------------------------- The Testing Set is : --------------------------------------------\n")
    print(test)
    print(f"\nThe Number of Enteries : {len(test)}\n")
    return train, test

# Transform data for neural network
def apply_transform(data, n: int):
    middle_data = []
    target_data = []
    for i in range(n, len(data)):
        input_sequence = data[i - n:i]
        middle_data.append(input_sequence)
        target_data.append(data[i])
    middle_data = np.array(middle_data).reshape((len(middle_data), n, 1))
    target_data = np.array(target_data)
    return middle_data, target_data

# LSTM model training function
def LSTM(train, n: int, number_nodes, learning_rate, epochs, batch_size):
    middle_data, target_data = apply_transform(train, n)  # Prepare the input data

    model = tf.keras.Sequential([
        tf.keras.layers.Input(shape=(n, 1)),  # Define the input shape here
        tf.keras.layers.LSTM(number_nodes, return_sequences=False),  # LSTM layer
        tf.keras.layers.Dense(units=number_nodes, activation="relu"),  # Fully connected layer
        tf.keras.layers.Dense(units=number_nodes, activation="relu"),  # Fully connected layer
        tf.keras.layers.Dense(1)  # Output layer for single value prediction
    ])

    model.compile(loss='mse', optimizer=tf.keras.optimizers.Adam(learning_rate), metrics=["mean_absolute_error"])

    # Debugging shapes
    print(f"middle_data shape: {middle_data.shape}")
    print(f"target_data shape: {target_data.shape}")

    # Fit the model
    history = model.fit(middle_data, target_data, epochs=epochs, batch_size=batch_size, verbose=0)

    # Get the full predictions on the training set
    full_predictions = model.predict(middle_data).flatten()

    # Print the model summary instead of accessing `input_shape` directly
    print("\nLSTM Model Summary:")
    model.summary()

    return model, history, full_predictions


# Calculate model accuracy
def calculate_accuracy(true_values, predictions):
    mse = mean_squared_error(true_values, predictions)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(true_values, predictions)
    return mse, rmse, mae

# Error evaluation
def Error_Evaluation(train_data, predict_train_data, n: int):
    errors = []
    for i in range(len(predict_train_data)):
        err = train_data[n + i] - predict_train_data[i]
        errors.append(err)
    return errors

# ARIMA parameter selection
def Parameter_calculation(data):
    finding = auto_arima(data, trace=True)
    plot_acf(data, lags=lag)
    address = os.path.join(Output_address, "ACF.jpg")
    plt.savefig(address)
    plot_pacf(data, lags=lag)
    address = os.path.join(Output_address, "PACF.jpg")
    plt.savefig(address)
    ord = finding.order
    return ord

# ARIMA model for predicting errors
def ARIMA_Model(train, len_test, ord):
    model = ARIMA(train, order=ord)
    model = model.fit()
    predictions = model.predict(start=len(train), end=len(train) + len_test, type='levels')
    full_predictions = model.predict(start=0, end=len(train) - 1, type='levels')
    return model, predictions, full_predictions

# Final predictions
def Final_Predictions(predictions_errors, predictions):
    final_values = []
    for i in range(days):
        final_values.append(predictions_errors[i] + predictions[i])
    return final_values

# Main function
def main():
    data = data_loader()
    plot_raw_data(data)
    train, test = data_allocation(data)
    plot_train_test(train, test)

    # LSTM Model
    st1 = time.time()
    model, history, full_predictions = LSTM(train, n, number_nodes, learning_rate, epochs, batch_size)
    plot_predictions(train[n:], full_predictions, "LSTM PREDICTIONS VS ACTUAL Values For TRAIN Data Set")

    # Predict future values with LSTM
    last_sequence = train[-n:].values.reshape((1, n, 1))
    predictions = []
    for i in range(days + 1):
        next_prediction = model.predict(last_sequence).flatten()[0]
        predictions.append(next_prediction)
        if i < len(test):
            actual_value = test.iloc[i]
            new_row = np.append(last_sequence[:, 1:, :], np.array([[[actual_value]]]), axis=1)
        else:
            new_row = np.append(last_sequence[:, 1:, :], np.array([[[next_prediction]]]), axis=1)
        last_sequence = new_row.reshape((1, n, 1))

    plot_predictions(test, predictions[:-1], "LSTM Predictions VS Actual Values")

    # Evaluate errors and plot them
    errors_data = Error_Evaluation(train, full_predictions, n)
    plot_prediction_errors(errors_data)

    mse, rmse, mae = calculate_accuracy(test[:days], predictions[:days])
    plot_accuracy(mse, rmse, mae)

    # ARIMA model for error correction
    ord = Parameter_calculation(errors_data)
    Arima_Model, predictions_errors, full_predictions_errors = ARIMA_Model(errors_data, len(test), ord)
    final_predictions = Final_Predictions(predictions_errors, predictions)
    plot_final_predictions(test[:days], final_predictions[:days])

if __name__ == '__main__':
    main()
```

1. Loading Libraries and Environment Variables

What we did: Imported necessary libraries for data manipulation (pandas, numpy), plotting (matplotlib), neural networks (tensorflow), time series modeling (statsmodels, pmdarima), and environmental variable management (dotenv).
Why we did it: These libraries are essential for loading, processing, visualizing data, and creating predictive models. Environment variables are loaded to store configurable parameters, making the code more flexible.
How we did it: We used the load_dotenv() method to read environment variables from a .env file and used default values in case they are missing.


2. Data Loading (data_loader)

What we did: Loaded the data from a CSV file, dropped missing values, and printed the shape of the dataset.
Why we did it: Data is required to train and test machine learning models. Cleaning the data ensures that missing values do not cause errors during model training.
How we did it: Used pandas.read_csv() to load the CSV and used dropna() to handle missing values.


3. Plotting Functions

What we did: Defined various plotting functions to visualize data, model predictions, errors, and accuracy metrics.
Why we did it: Visualizing data helps in understanding the model’s performance, comparing predicted vs. actual values, and debugging any errors.
How we did it: Used matplotlib.pyplot to create and save plots of actual vs. predicted values, errors, and accuracy metrics like MSE, RMSE, and MAE.


4. Data Partitioning (data_allocation)

What we did: Split the dataset into a training set and a testing set.
Why we did it: Machine learning models require training data to learn from and testing data to evaluate their performance on unseen data. Splitting helps measure generalization.
How we did it: Used .iloc[] to select the first portion of the dataset for training and the last portion for testing based on the days variable.


5. Data Transformation (apply_transform)

What we did: Transformed the data into a format suitable for LSTM, where the input is a sequence of n past values and the target is the next value.
Why we did it: LSTMs require input in a sequence format (lagged observations) to predict the next value in a time series.
How we did it: Created sequences of data by taking n consecutive observations and reshaping the data using numpy.


6. LSTM Model Training (LSTM)

What we did: Built an LSTM model with 2 hidden layers (Dense layers), compiled it with Adam optimizer and mse loss, and trained it on the transformed training data.
Why we did it: LSTMs are commonly used for time series forecasting because they can capture long-term dependencies in sequential data.
How we did it: Used tensorflow.keras.Sequential() to build the model and used fit() to train it.


7. Model Evaluation and Prediction

What we did: Made predictions on the training data and the test set, visualized the actual vs. predicted values, and calculated the mean squared error (MSE), root mean squared error (RMSE), and mean absolute error (MAE).
Why we did it: Predictions help evaluate the performance of the model. MSE, RMSE, and MAE are standard metrics for regression problems, measuring how close the predictions are to the actual values.
How we did it: Used model.predict() to generate predictions and calculated metrics using mean_squared_error() and mean_absolute_error().


8. ARIMA Error Correction Model

What we did: Built an ARIMA model to correct errors from the LSTM model by modeling the residuals (errors). The ARIMA model predicts errors which are then added to the LSTM predictions.
Why we did it: Hybrid models like LSTM + ARIMA can improve the accuracy of predictions by combining the strengths of both models. LSTM captures the long-term dependencies, while ARIMA models the residuals.
How we did it: Used auto_arima() to automatically select the best ARIMA parameters and then trained an ARIMA model on the residuals.


9. Final Predictions

What we did: Combined the LSTM predictions with ARIMA error corrections and visualized the final predictions.
Why we did it: The final predictions reflect both the sequential dependencies captured by LSTM and the error correction by ARIMA.
How we did it: Added the ARIMA-predicted errors to the LSTM predictions and visualized the results.
